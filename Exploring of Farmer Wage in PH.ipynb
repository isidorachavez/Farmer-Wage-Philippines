## Packages

```python
#importing important packages
import pandas as pd
```

## Loading 

```python
#viewing of dataset
df = pd.read_excel('FarmerWageCleaned_v1.xlsx')
```

## Cleaning

```python
df.head(10)
```

```python
#Creating list for new column names
#Start year
x = 1975

#Range of years from 1975 to 2018
y = 2018-1974

#Creating list for column names
z = ['Wage Type', 'Farmer Type','Location'] 
for n in range(y):
    z.extend([x+n for i in range(3)]) 
```

```python
#Renaming of columns
df.columns = z
```

```python
#Making first 3 cells of row 1 to blank
df.iloc[0,0] = ""
df.iloc[0,1] = ""
df.iloc[0,2] = ""
```

```python
#Keeping only necessary non-NA columns
df = df[df['Location'].notna()]
```
